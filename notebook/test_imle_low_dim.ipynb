```python
import imle
```

```python
noise_value=0.3
```

```python
model = imle.Imle(in_ndims=1, out_ndims=1, alpha=0.995,Psi0=[noise_value**2] * 1, sigma0=pi/2, wsigma=1**1, wSigma=1**1, wNu=1,wLambda=0.1, wPsi=1, p0=0.1, multiValuedSignificance=0.8, nSolMax=6)
```

```python
z_train=arange(-2*pi, 2*pi, 0.1)
z_train.shape=(max(z_train.shape),1)
```

```python
x_train=cos(z_train)+noise_value*rand(*z_train.shape)
```

```python
data=hstack((z_train, x_train))
shuffle(data)
z_train=data[:,0:1]
x_train=data[:,1:2]
z_train.shape
```

```python
for i, (z, x) in enumerate(zip(z_train, x_train)):
    if i % 1000 == 0:
        print i
    
    model.update(z, x)
```

```python
print model.number_of_experts
```

```python
mu=array([model.get_joint_mu(e) for e in range(model.number_of_experts)])
mu.shape
```

```python
COLORS = ['r', 'g', 'b', 'k', 'm']*10
```

```python
plot(z_train.T,x_train.T, 'xb')
plot(mu[:,0], mu[:,1],'ob')
a=subplot(111)
gmm=model.to_gmm()
els = gmm.get_display_ellipses2D(COLORS)
for el in els:
    a.add_patch(el)
for goal in arange(-8,3,0.05):
    sols=model.predict([goal])
    #plot(tile(goal,[sols.shape[0],1]), sols, 'or')
```
