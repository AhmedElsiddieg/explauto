```python
from explauto.experiment import Experiment, make_settings

s = make_settings(environment='musical_arm',
                  babbling_mode='goal', 
                  interest_model='discretized_progress',
                  sensorimotor_model='nearest_neighbor')

expe = Experiment.from_settings(s)
```

```python
%pylab inline
ax = axes()
expe.env.arm.plot_arm(ax, expe.env.arm.random_motors()[0])
expe.env.plot_keyboard(ax)
```

```python
expe.run(1000)
```

```python
%pylab
clf()
last_t = 0
for t in linspace(100, 1000, 100):
    t = int(t)
    ax = subplot(121)
    expe.log.scatter_plot(ax, (('sensori', [0, 1]),), range(0, t), marker='.', markersize=0.3, color = 'white')
    expe.log.density_plot(ax, (('choice', [0, 1]),), range(last_t, t))
    expe.env.plot_keyboard(ax)
    ax = subplot(122)
    expe.log.scatter_plot(ax, (('sensori', [1, 2]),), range(0, t), False, marker='.', markersize=0.3, color = 'white')
    expe.log.density_plot(ax, (('choice', [1, 2]),), range(last_t, t))
    #ax.axis('tight')
    draw()
    last_t = t
```

```python
%pylab inline
figure()
ax = subplot(121)
expe.log.density_plot(ax, (('sensori', [0, 1]),))
expe.log.scatter_plot(ax, (('sensori', [0, 1]),), marker='.', markersize=1, color = 'white')
expe.env.plot_keyboard(ax)
ax = subplot(122)
expe.log.density_plot(ax, (('sensori', [1, 2]),))
expe.log.scatter_plot(ax, (('sensori', [1, 2]),), marker='.', markersize=1, color = 'white')
```
